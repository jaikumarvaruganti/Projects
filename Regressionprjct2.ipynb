```python
# Project Title: STUDENT SUCCESS PREDICTION (Using Linear Regression)

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv(r"C:\Users\vmahe\Downloads\StudentsPerformanceDataset1.csv")

X = data[['Study_Hours']]
y = data[['Final_Score']]

model= LinearRegression()
model.fit(X,y)
predicted_scores = model.predict(X)


mae = mean_absolute_error(y,predicted_scores)
mse = mean_squared_error(y,predicted_scores)
rmse =np.sqrt(mse)
r2 = r2_score(y,predicted_scores)

print("Mean Absolute Error:",round(mae,2))
print("Mean Squared Error:", round(mse,2))
print("Root Mean Squared Error:", round(rmse,2))
print("R^2 Score (model accuracy):",round(r2,2))

# HISTOGRAM
plt.figure(figsize= (10, 8))
plt.hist(data['Final_Score'], bins = 40, color = 'skyblue', edgecolor= 'black')
plt.title("Ditribution of Final Exam Score")
plt.xlabel("Final Exam Score")
plt.ylabel("No of Students")
plt.grid()
plt.show()

# Scatter And Regression Plot
plt.figure(figsize= (10, 8))
plt.scatter(X,y, color ='blue', label = 'Actual Scores')
plt.plot(X, predicted_scores, color= 'red', label = 'Predicted Scores( Regression Line)')
plt.title("Model Prediction VS Actual Score")
plt.xlabel("Study Hours Per Week")
plt.ylabel("Final Output")
plt.grid()
plt.show()

# Correlation Matrix
columns = ['Study_Hours','Final_Score','Midterm_Score','Assignments_Avg','Quizzes_Avg','Participation_Score','Projects_Score']
corr_matrix = data[columns].corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',linewidth = '1', linecolor= 'black')
plt.title("Correlation Heatmap")
plt.show()

new_hours = float(input("Enter the number of hours studied:"))
predicted_new_score= model.predict([[new_hours]])
print(f"If they study for {new_hours} Hours/week, they will get {predicted_new_score} marks.")

""" Conclusion:-
Our linear regression model explains 89% of the variation in exam scores based on study hours.
On average, the predictions are within 67 points of the actual scores,
which makes the model highly accurate for this dataset.
"The Relationship is Positive And Strong(as Hours increase Score also increase"
 *It says that the more they study, the better their Exam Score will be.     """
```
